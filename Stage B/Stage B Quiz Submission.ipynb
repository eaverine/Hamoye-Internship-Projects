```python
import pandas as pd

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv'
df = pd.read_csv(url)
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler

new_df = df.drop(columns = ['date', 'lights'])
scaler = MinMaxScaler()
normalized_new_df = pd.DataFrame(scaler.fit_transform(new_df), columns = new_df.columns)
```

```python
target_variable = normalized_new_df['Appliances']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


linear_model = LinearRegression()
x_train, x_test, y_train, y_test = train_test_split(normalized_new_df, target_variable, test_size = 0.3, random_state = 42)
linear_model.fit(x_train, y_train)
```

```python
predicted_output = linear_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score

linear_df_r2_score = r2_score(predicted_output, y_test)
round(linear_df_r2_score, 2)
```

```python
from sklearn.metrics import mean_absolute_error

linear_df_mae = mean_absolute_error(predicted_output, y_test)
round(linear_df_mae, 2)
```

```python
import numpy as np

linear_df_rss = np.sum(np.square(predicted_output - y_test))
round(linear_df_rss, 2)
```

```python
from sklearn.metrics import mean_squared_error as mse

linear_df_rmse = np.sqrt(mse(predicted_output, y_test))
round(linear_df_rmse, 3)
```

```python
from sklearn.linear_model import Lasso

lasso_reg = Lasso()
lasso_reg.fit(x_train, y_train)
```

```python
lasso_predicted_output = lasso_reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

lasso_r2_score = r2_score(lasso_predicted_output, y_test)
print('Lasso r2 score: ', round(lasso_r2_score, 2))

lasso_mae = mean_absolute_error(lasso_predicted_output, y_test)
print('Lasso mean absolute error: ', round(lasso_mae, 2))

lasso_rss = np.sum(np.square(lasso_predicted_output - y_test))
print('Lasso residual sum of squares: ', round(lasso_rss, 2))

lasso_rmse = np.sqrt(mean_squared_error(lasso_predicted_output, y_test))
print('Lasso root mean squared error: ', round(lasso_rmse, 2))
```

```python
from sklearn.linear_model import Ridge

ridge_reg = Ridge()
ridge_reg.fit(x_train, y_train)
```

```python
ridge_predicted_output = ridge_reg.predict(x_test)
```

```python
ridge_r2_score = r2_score(ridge_predicted_output, y_test)
print('Ridge r2 score: ', round(ridge_r2_score, 2))

ridge_mae = mean_absolute_error(ridge_predicted_output, y_test)
print('Ridge mean absolute error: ', round(ridge_mae, 2))

ridge_rss = np.sum(np.square(ridge_predicted_output - y_test))
print('Ridge residual sum of squares: ', round(ridge_rss, 2))

ridge_rmse = np.sqrt(mean_squared_error(ridge_predicted_output, y_test))
print('Ridge root mean squared error: ', round(ridge_rmse, 2))
```
